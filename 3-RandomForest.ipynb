```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
```

```python
import nltk
#nltk.download('stopwords')
from nltk import word_tokenize
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
text = pd.read_csv("data/constitutions.csv")
text.head(5)
```

```python
import string

def text_cleaning(text):
    remove_punctuation = [char for char in text if char not in string.punctuation]
    remove_punctuation = "".join(remove_punctuation)  
    return [word for word in remove_punctuation.split() if word.lower() not in stopwords.words('english')]
```

```python
print(text['content'].apply(text_cleaning))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_transformer = CountVectorizer(analyzer=text_cleaning).fit(text['content'])
content_bow = bow_transformer.transform(text['content'])
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer().fit(content_bow)
content_tfidf = tfidf_transformer.transform(content_bow)
```

```python
labels = pd.read_csv("data/labels.csv")
labels.head()
```

```python
display(labels.iloc[:,40])
```

```python
from sklearn.model_selection import train_test_split
for i in range(0,40):
    X_train, X_test, y_train, y_test = train_test_split(content_tfidf, labels.iloc[:,i], test_size=0.3)
    
```
