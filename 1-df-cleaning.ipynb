```python
import pandas as pd

df = pd.read_csv("data/constitutions.csv")
df['country'] = ""
df['year'] = ""
df['latest rev'] = ""
```

```python
import re

for index, row in df.iterrows():
    patt1 = re.findall('([a-zA-Z ]*)\d*.*', row['title'])
    row['country'] = str(patt1[0]).replace(" ", "-")[:-1].lower()
    
    patt2 = re.findall(r'\d+', row['title'])
    res = list(map(int, patt2))
    row['year'] = int(res[0])
    row['latest rev'] = int(res[-1])
    row['content'] = row['content'].lower()
```

```python
df.sort_values(by='country', ascending=True, inplace=True)
df = df[df.country != ""]
df
```
